## Bucles

> For, es una estructura de iteración, recorre todo el elemento 


```python
learners = [
    {"nombre": "Pepita", "promedio": 8.7},
    {"nombre": "Pepito", "promedio": 7.5},
    {"nombre": "Andres", "promedio": 2},
    {"nombre": "Juan", "promedio": 8.7},
    {"nombre": "Cesar", "promedio": 3.5},
    {"nombre": "Carlos", "promedio": 3.5},
    {"nombre": "Cami", "promedio": 2},
    {"nombre": "Pepito", "promedio": 7.5},
    {"nombre": "Andres", "promedio": 9.1},
    {"nombre": "Pepita", "promedio": 2},
    {"nombre": "Pepito", "promedio": 7.5},
    {"nombre": "Andres", "promedio": 3.5},
    {"nombre": "Pepita", "promedio": 8.7},
    {"nombre": "Pepito", "promedio": 7.5},
    {"nombre": "Andres", "promedio": 5},
]

for learner in learners:
    if learner["promedio"] >= 7:
        print(f"Felicitaiones {learner['nombre']}")
    elif learner["promedio"] >=5:
        print(f"{learner['nombre']} Puedes mejorar")
    else:
        print(f"Necesitas mejorar")
```
