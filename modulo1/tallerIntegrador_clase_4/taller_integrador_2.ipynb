## Grafica 1

```python
import matplotlib.pyplot as plt
```

```python
supervivencia_sexo = titanic_limpio.groupby("sexo")["sobrevivio"].mean()

plt.figure(figsize =(7,5))
supervivencia_sexo.plot(
    kind="bar",
    color = ["#DD8452","#3062B8"],
    edgecolor ="white",
    width=0.5
)

plt.title("supervivencia por sexo", fontsize=14, fontweight="bold")
plt.xlabel("")
```

```python
supervivencia_clase = titanic_limpio.groupby("clase")["sobrevivio"].mean()

plt.figure(figsize =(7,5))
supervivencia_clase.plot(
    kind="bar",
    color = ["#D21414","#3062B8"],
    edgecolor ="white",
    width=0.5
)

plt.title("supervivencia por clase", fontsize=14, fontweight="bold")
plt.xlabel("")
```
